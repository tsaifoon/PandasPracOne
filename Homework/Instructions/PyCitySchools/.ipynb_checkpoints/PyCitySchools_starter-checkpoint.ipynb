```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
tot_school = len(school_data["school_name"].unique())
# Calculate the total number of students
tot_students = student_data["Student ID"].count()
# Calculate the total budget
tot_budget = school_data["budget"].sum()
# Calculate the average math score
avg_math = student_data["math_score"].mean()
# Calculate the average reading score
avg_read = student_data["reading_score"].mean()
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
ove_avg_score = (avg_math + avg_read) / 2
# Calculate the percentage of students with a passing math score (70 or greater)
per_pass_math = student_data.loc[student_data["math_score"] > 70,"math_score"].count() / tot_students * 100
# Calculate the percentage of students with a passing reading score (70 or greater)
per_pass_read = student_data.loc[student_data["reading_score"] > 70,"reading_score"].count() / tot_students * 100
# Create a dataframe to hold the above results
result_DF = pd.DataFrame({
    "Total School" : [tot_school],
    "Total Students" : [tot_students],
    "Total Budget" : [tot_budget],
    "Average Math" : [avg_math],
    "Average Reading" : [avg_read],
    "Overall Average Score" : [ove_avg_score],
    "Percent Pass Math" : [per_pass_math],
    "Percent Pass Reading" : [per_pass_read]})
result_DF
# Optional: give the displayed data cleaner formatting

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school, including:
gb_sch_name = school_data_complete.groupby("school_name")

# School Name
uni_sch_name = gb_sch_name["school_name"].unique()

# School Type
uni_sch_type = gb_sch_name["type"].unique()

# Total Students
uni_tot_stud = gb_sch_name["Student ID"].count()

# Total School Budget
uni_tot_budg = gb_sch_name["budget"].unique()

# Per Student Budget
per_stud_budg = uni_tot_budg / uni_tot_stud

# Average Math Score
uni_avg_math = gb_sch_name["math_score"].mean()

# Average Reading Score
uni_avg_read = gb_sch_name["reading_score"].mean()

# % Passing Math
math_org = school_data_complete.loc[school_data_complete["math_score"]>70,:]
math_df = math_org.groupby("school_name")
uni_pass_math = math_df["math_score"].count() / uni_tot_stud * 100

# % Passing Reading
read_org = school_data_complete.loc[school_data_complete["reading_score"]>70,:]
read_df = read_org.groupby("school_name")
uni_pass_read = read_df["reading_score"].count() / uni_tot_stud * 100

# Overall Passing Rate (Average of the above two)
over_pass_rate = (uni_pass_math + uni_pass_read)/2

#Create a dataframe to hold the above results
by_sch_data = pd.DataFrame({
    "School Name": uni_sch_name.str[0],
    "School Type": uni_sch_type.str[0],
    "Total Student": uni_tot_stud,
    "Total School Budget": uni_tot_budg.str[0],
    "Per Student Budget": per_stud_budg.str[0],
    "Average Math Score": uni_avg_math,
    "Average Reading Score": uni_avg_read,
    "Passing Math": uni_pass_math,
    "Passing Reading": uni_pass_read,
    "Overall Passing Rate": over_pass_rate
})
by_sch_data = by_sch_data.set_index("School Name")
by_sch_data
```

## Top Performing Schools (By Passing Rate)

```python
#Sort and display the top five schools in overall passing rate
top_per_sch = by_sch_data.sort_values(["Overall Passing Rate"],ascending=False)
top_per_sch.reset_index(inplace=True)
top_per_sch.head(5)

```
